```python
import os

from genomics.popgen.plink.convert import to_eigen
from genomics.popgen.pca import plot, smart
%matplotlib inline
```

## Requires plink from data preparation

```python
f = open('relationships_w_pops_121708.txt')
ind_pop = {}
f.readline()  # header
for l in f:
    toks = l.rstrip().split('\t')
    fam_id = toks[0]
    ind_id = toks[1]
    pop = toks[-1]
    ind_pop['/'.join([fam_id, ind_id])] = pop
ind_pop['2469/NA20281'] = ind_pop['2805/NA20281'] 
```

```python
os.system('plink --recode12 --file hapmap10 --noweb --out hapmap10_12')
# move this to data management
```

```python
to_eigen('hapmap10_12', 'hapmap10')
```

## Running smartpca

```python
ctrl = smart.SmartPCAController('hapmap10')
ctrl.run()
```

```python
wei, wei_perc, ind_comp = smart.parse_evec('hapmap10.evec', 'hapmap10.eval')
```

```python
%matplotlib inline
plot.render_pca(ind_comp, 1, 2, cluster=ind_pop)
```

```python
size = 50
step = 10
os.system("plink --silent --file hapmap10 --indep-pairwise %d %d 0.1 --out SNPStoBeKept --noweb" % (size, step))
```

```python
os.system('plink --recode12 --file hapmap10 --noweb --extract SNPStoBeKept.prune.in  --out hapmap10_ld_12')
```
