```python
from collections import defaultdict
import functools

%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt

import vcf
```

```python
def do_window(recs, size, fun):
    start = None
    win_res = []
    for rec in recs:
        if not rec.is_snp or len(rec.ALT) > 1:
            continue
        if start is None:
            start = rec.POS
        my_win = 1 + (rec.POS - start) // size
        while len(win_res) < my_win:
            win_res.append([])
        win_res[my_win - 1].extend(fun(rec))
    return win_res
```

```python
def apply_win_funs(wins, funs):
    fun_results = []
    for win in wins:
        my_funs = {}
        for name, fun in funs.items():
            try:
                my_funs[name] = fun(win)
            except:
                my_funs[name] = None
        fun_results.append(my_funs)
    return fun_results
```

```python
wins = {}
size = 2000
for name in ['centro.vcf.gz', 'standard.vcf.gz']:
    recs = vcf.Reader(filename=name)
    wins[name.split('.')[0]] = do_window(recs, size, lambda x: [1])
```

```python
stats = {}
fig, ax = plt.subplots(figsize=(16, 9))
for name, nwins in wins.items():
    stats[name] = apply_win_funs(nwins, {'sum': sum})
    ax.plot([x['sum'] for x in stats[name]], label=name)
ax.legend()
```

```python
mq0_wins = {}
size = 5000
def get_sample(rec, annot, my_type):
    #We will ignore Nones
    res = []
    samples = rec.samples
    for sample in samples:
        if sample[annot] is None:
            continue
        res.append(my_type(sample[annot]))
    return res

for name in ['centro.vcf.gz', 'standard.vcf.gz']:
    recs = vcf.Reader(filename=name)
    mq0_wins[name.split('.')[0]] = do_window(recs, size, functools.partial(get_sample, annot='MQ0', my_type=int))
```

```python
import numpy as np
stats = {}
colors = ['b', 'g']
i = 0
fig, ax = plt.subplots(figsize=(16, 9))
for name, nwins in mq0_wins.items():
    stats[name] = apply_win_funs(nwins, {'median': np.median, '75': functools.partial(np.percentile, q=95)})
    ax.plot([x['median'] for x in stats[name]], label=name, color=colors[i])
    ax.plot([x['75'] for x in stats[name]], '--', color=colors[i])
    i += 1
#ax.set_ylim(0, 40)
plt.legend()
```

```python
def get_sample_correlation(recs, a1, a2, t1=int, t2=int):
    correl = defaultdict(int)
    for rec in recs:
        if not rec.is_snp:
             continue
        for sample in rec.samples:
            if sample[a1] is None or sample[a2] is None:
                continue  # We ignore Nones
            correl[(t1(sample[a1]), t2(sample[a2]))] += 1
            #careful with the size, floats: round?
    return correl
```

```python
name = 'standard.vcf.gz'
recs = vcf.Reader(filename=name)

correl = get_sample_correlation(recs, 'GT', 'DP', t1=str) # There might be better ways of extracting GL
```

```python
dps = [x[1] for x in correl.keys()]
dps.sort()
frac_hz = []
cnt_dp = []
fig, ax = plt.subplots(figsize=(16, 9))
for dp in dps:
    hz = 0.0
    cnt = 0
    
    for gt, kdp in correl.keys():
        if kdp != dp:
            continue
        cnt += correl[(gt, dp)]
        if gt[0] != gt[2]:
            hz += correl[(gt, dp)]
    frac_hz.append(hz / cnt)
    cnt_dp.append(cnt)
ax.plot(dps, frac_hz)
ax2 = ax.twinx()
ax2.plot(dps, cnt_dp, 'r')
ax.set_xlim(0, 75)
ax.set_ylim(0, 0.2)
ax2.set_ylabel('Quantity of calls')
ax.set_ylabel('Fraction of Heterozygote calls')
ax.set_xlabel('Sample Read Depth (DP)')
fig.suptitle('Number of calls per depth and fraction of that are heterozyguous',fontsize='xx-large')
```
