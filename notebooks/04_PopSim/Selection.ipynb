```python
from collections import OrderedDict
from copy import deepcopy

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import simuPOP as sp
%matplotlib inline
```

```python
num_loci = 10
pop_size = 1000
num_gens = 101
```

```python
init_ops = OrderedDict()
pre_ops = OrderedDict()
post_ops = OrderedDict()
```

```python
pops = sp.Population(pop_size, loci=[1 * num_loci], infoFields=['fitness'])
```

```python
def create_derived_by_count(pop, param):
    #Assumes everything is autosomal and that derived is low (<0.5)
    locus, cnt = param
    for i, ind in enumerate(pop.individuals()):
        for marker in range(pop.totNumLoci()):
            if i < cnt and locus == marker:
                ind.setAllele(1, marker, 0)
            else:
                ind.setAllele(0, marker, 0)
            ind.setAllele(0, marker, 1)
    return True
```

```python
init_ops['Sex'] = sp.InitSex()
init_ops['Freq-sel'] = sp.PyOperator(create_derived_by_count, param=(0, 10))
#careful above
init_ops['Freq-neutral'] = sp.InitGenotype(freq=[0.5, 0.5], loci=range(1, num_loci))
#init_ops['ble'] = sp.PyExec('ble = []')
#post_ops['ble'] = sp.PyEval(r'" %s\n" % str(ble)')
#pre_ops['ble'] = sp.PyExec('ble.append(gen)')
post_ops['Stat-freq'] = sp.Stat(alleleFreq=sp.ALL_AVAIL)
post_ops['Stat-freq-eval'] = sp.PyEval(r"'%d %.3f\n' % (gen, alleleFreq[0][1])", reps=[0], step=10)
```

```python
ms = sp.MapSelector(loci=0, fitness={
                (0, 0): 0.90,
                (0, 1): 1,
                (1, 1): 1})
mating_scheme = sp.RandomMating(ops=[sp.MendelianGenoTransmitter(), ms])
```

```python
def init_acumulators(pop, param):
    acumulators = param
    for acumulator in acumulators:
        pop.vars()[acumulator] = []
    return True

def update_acumulator(pop, param):
    acumulator, var = param
    pop.vars()[acumulator].append(deepcopy(pop.vars()[var]))
    return True
```

```python
def get_freq_deriv(pop, param):
    marker = param
    expHe = {}
    pop.dvars().freqDeriv = pop.dvars().alleleFreq[0][1]
    return True
```

```python
init_ops['acumulators'] = sp.PyOperator(init_acumulators, param=['freq_sel'])
post_ops['FreqSel'] = sp.PyOperator(get_freq_deriv, param=0)
post_ops['freq_sel_acumulation'] = sp.PyOperator(update_acumulator, param=('freq_sel', 'freqDeriv'))
```

```python
sim = sp.Simulator(pops, rep=100)  # talk about threads
sim.evolve(initOps=init_ops.values(), preOps=pre_ops.values(), postOps=post_ops.values(),
           matingScheme=mating_scheme, gen=num_gens)
None
```

```python
fig = plt.figure(figsize=(16, 9))
ax = fig.add_subplot(111)
for pop in sim.populations():
    ax.plot(pop.vars()['freq_sel'])
```

```python
def calc_exp_he(pop):
    #assuming bi-allelic markers coded as 0 and 1
    #sample size not corrected - OK
    pop.dvars().expHe = {}
    for locus, freqs in pop.dvars().alleleFreq.items():
        f0 = freqs[0]
        pop.dvars().expHe[locus] = 1 - f0**2 - (1 - f0)**2
    return True
```

```python
init_ops['acumulators'] = sp.PyOperator(init_acumulators, param=['num_males', 'exp_he'])
post_ops['Stat-males'] = sp.Stat(numOfMales=True)
post_ops['ExpHe'] = sp.PyOperator(calc_exp_he)
post_ops['male_acumulation'] = sp.PyOperator(update_acumulator, param=('num_males', 'numOfMales'))
post_ops['expHe_acumulation'] = sp.PyOperator(update_acumulator, param=('exp_he', 'expHe'))
#remember ordering
#generation step
```

```python
del post_ops['Stat-freq-eval']
```

```python
num_gens = 100
pops_500 = sp.Population(500, loci=[1 * num_loci])
sim = sp.Simulator(pops_500, rep=1)
sim.evolve(initOps=init_ops.values(), preOps=pre_ops.values(), postOps=post_ops.values(),
           matingScheme=mating_scheme, gen=num_gens)
```

```python
pop_500_after = deepcopy(sim.population(0))
```

```python
pops_40 = sp.Population(40, loci=[1 * num_loci])
sim = sp.Simulator(pops_40, rep=1)
sim.evolve(initOps=init_ops.values(), preOps=pre_ops.values(), postOps=post_ops.values(),
           matingScheme=mating_scheme, gen=num_gens)
```

```python
pop_40_after = deepcopy(sim.population(0))
```

```python
def calc_loci_stat(var, fun):
    stat = []
    for gen_data in var:
        stat.append(fun(gen_data.values()))
    return stat
```

```python
sns.set_style('white')
fig, axs = plt.subplots(1, 2, figsize=(16, 9), sharey=True, squeeze=False)
def plot_pop(ax1, pop):
    for locus in range(num_loci):
        ax1.plot([x[locus] for x in pop.dvars().exp_he], color=(0.75, 0.75, 0.75))
    mean_exp_he = calc_loci_stat(pop.dvars().exp_he, np.mean)
    ax1.plot(mean_exp_he, color='r')

plot_pop(axs[0, 0], pop_40_after)
plot_pop(axs[0, 1], pop_500_after)
ax = fig.add_subplot(4, 4, 13)
ax.boxplot(pop_40_after.dvars().num_males)
ax = fig.add_subplot(4, 4, 16)
ax.boxplot(pop_500_after.dvars().num_males)
fig.tight_layout()
```
